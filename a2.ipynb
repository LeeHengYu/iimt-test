```python
# Q1

from math import inf 
def check_int(s):
    if s[0] == '-':
        return s[1:].isdigit()
    return s.isdigit()

def repeat_input():
    flag = False
    min_num = inf
    max_num = -inf
    while True:
        msg = input('Enter an integer number or done: ')
        if msg == 'done':
            if flag:
                break
            else:
                print('***WARNING: you have not entered a number yet. ***')
        elif not check_int(msg):
            print('***REMINDER: please enter an integer number or done***')
        else:
            min_num = min(min_num, int(msg))
            max_num = max(max_num, int(msg))
            flag = True
    print() # empty line
    print(f'The smallest number you entered is: {min_num}')
    print(f'The biggest number you entered is: {max_num}')
    
repeat_input()
```

```python
# Q2
from random import randint

def turtle_rabbit(n):
    rabbit_loc = 0
    turtle_loc = 0
    for i in range(n):
        if i % 5 != 4:
            rabbit_loc += randint(4, 8)
        turtle_loc += randint(2, 6)
        print(f'Round {i+1}')
        print('.'*turtle_loc, 'Turtle', f'({turtle_loc})', sep=' ')
        print('.'*rabbit_loc, 'Rabbit', f'({rabbit_loc})', sep=' ')
        print() # empty line
    if turtle_loc == rabbit_loc:
        print("It's a tie.")
    else:
        print(f'{'Turtle' if turtle_loc > rabbit_loc else 'Rabbit'} wins!')
        
turtle_rabbit(12)
```

```python
# Q3

'''
Intuition: if we can use for loop or while loop, we can variably print out n1 numbers of lines based on the input n1.
If we can't use for loop or while loop, we need to put a fixed number of statements related to print, which do not necessarily print out contents.
Ideally it can be a function that is responsible for printing, but has the ability not to print if the input is not desired.
Practice: use for loop and while loop to practice this question
'''

def print_line(n1, n2):
    if n1 < 1: # invalid input
        return
    print(str(n1)*n1, str(n2-n1)*(n2-n1), sep=' ')

def number_pattern(n1, n2):
    if type(n1) != int or type(n2) != int or min(n1, n2) < 1 or max(n1, n2) > 9 or n1 >= n2:
        print('Please make sure the two arguments are integers between 1 and 9, and that the 2nd argument is greater than the 1st one.')
        return
    print_line(n1, n2)
    print_line(n1-1, n2)
    print_line(n1-2, n2)
    print_line(n1-3, n2)
    print_line(n1-4, n2)
    print_line(n1-5, n2)
    print_line(n1-6, n2)
    print_line(n1-7, n2)
    print_line(n1-8, n2)
    # We have at most 9 lines to print
    
number_pattern(6, 1)
number_pattern(0, 7)
number_pattern(5, 12)
number_pattern('a', 'f')
number_pattern(1, 3)   
number_pattern(5, 8)   
number_pattern(8, 9)    
```

```python
# Q4

def check(n):
    '''Check if n satifies the condition'''
    return n == sum(int(i)**len(str(n)) for i in str(n))

def sum_of_powers(a, b):
    res = []
    for n in range(a, b+1):
        if check(n):
            res.append(n)
    for i in range(len(res)):
        if i != len(res)-1:
            print(res[i], end='  ')
        else:
            print(res[i]) 
            
sum_of_powers(100, 1500)
sum_of_powers(2000, 99999)
```
